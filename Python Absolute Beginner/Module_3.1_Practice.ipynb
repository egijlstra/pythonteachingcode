### Guessing a letter A-Z  
**check_guess()** takes 2 string arguments: **letter and guess** (both expect single alphabetical character)   
    - if guess is not an alpha character print invalid and return False
    - test and print if guess is "high" or "low" and return False
    - test and print if guess is "correct" and return True

```python
# [ ] create check_guess()
# call with test
def check_guess(letter, guess):
    if guess < letter:
        print("Your guess is too low.")
        return False
    elif guess > letter:
        print("Your guess is too high.")
        return False
    elif guess == letter:
        print('Correct')
        return True
    else: 
        guess.isalnum() == False
        print('invalid')
        return False
        
check_guess('A', 'F')
```

```python
# [ ] call check_guess with user input
correct_letter = input("Enter the secret letter (A-Z): ")
your_guess = input("Enter your guess (A-Z): ")

check_guess(correct_letter, your_guess)
```

### Letter Guess
**create letter_guess() function that gives user 3 guesses**
- takes a letter character argument for the answer letter
- gets user input for letter guess  
- calls check_guess() with answer and guess
- End letter_guess if 
    - check_guess() equals True, return True  
    - or after 3 failed attempts, return False

```python
# [ ] create letter_guess() function, call the function to test
def letter_guess(letter):
    attempts = 0
    while attempts < 3:
        guess = input("Enter your guess (A-Z): ")
        if check_guess(letter, guess):
            return True
        attempts += 1
    print("Out of attempts! The correct letter was:", letter)
    return False

letter_guess('D')
    

```

### Pet Conversation
**ask the user for a sentence about a pet and then reply**  
- get user input in variable: about_pet
- using a series of **if** statements respond with appropriate conversation
  - check if "dog" is in the string about_pet (sample reply "Ah, a dog")
  - check if "cat" is in the string about_pet
  - check if 1 or more animal is in string about_pet
- no need for **else**'s
- finish with thanking for the story

```python
# [ ] complete pet conversation
about_pet = input('Tell me a story about your pet(s): ')

if 'dog' in about_pet:
    print('Ah, a dog.')
if 'cat' in about_pet:
    print('Ah, a cat.')
animals = ["dog", "cat", "bird", "fish", "rabbit", "hamster", "turtle", "snake", "lizard", "horse"]
if any(animal in about_pet for animal in animals):
    print('You have many pets!')

print('Thanks for sharing that story!')
```

### Rainbow colors
ask for input of a favorite rainbow color first letter: ROYGBIV  

Using `if`, `elif`, and `else`:  
- print the color matching the letter  
    - R = Red  
    - O = Orange  
    - Y = Yellow  
    - G = Green
    - B = Blue
    - I = Indigo
    - V = Violet
    - else print "no match"


```python
# [ ] complete rainbow colors
letter = input('what is the first letter your favorite rainbow color (ROYGBIV)?')

if letter == "R":
    print("Red")
elif letter == "O":
    print("Orange")
elif letter == "Y":
    print("Yellow")
elif letter == "G":
    print("Green")
elif letter == "B":
    print("Blue")
elif letter == "I":
    print("Indigo")
elif letter == "V":
    print("Violet")
else:
    print("No match")
```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter, 
# get input and call the function and return the matching color as a string or "no match" message.
# Call the function and print the return string.
def rainbow_color(letter):
    letter = letter.upper()
    if letter == "R":
        return "Red"
    elif letter == "O":
        return "Orange"
    elif letter == "Y":
        return "Yellow"
    elif letter == "G":
        return "Green"
    elif letter == "B":
        return "Blue"
    elif letter == "I":
        return "Indigo"
    elif letter == "V":
        return "Violet"
    else:
        return "No match"

letter = input("Enter the first letter of your favorite rainbow color (ROYGBIV): ")
print(rainbow_color(letter))

```

**Create function age_20() that adds or subtracts 20 from your age for a return value based on current age** (use `if`) 
- call the funtion with user input and then use the return value in a sentence  
example `age_20(25)` returns **5**: 
> "5 years old, 20 years difference from now"

```python
# [ ] complete age_20()
def age_20(age):
    if age > 20:
        print(age - 20, 'years old, 20 years ago')
    elif age < 20:
        print(age + 20, 'years old, 20 years difference from now')
    else:
        print('invalid age')
        
age = int(input('How old are you?: '))
print(age_20(age))
```

**create a function rainbow_or_age that takes a string argument**
- if argument is a digit return the value of calling age_20() with the str value cast as **`int`** 
- if argument is an alphabetical character return the value of calling rainbow_color() with the str
- if neither return FALSE

```python
# [ ]  create rainbow_or_age()
def rainbow_or_age(argument):
    if argument.isdigit():
        return age_20(int(argument))
    elif argument.isalpha():
        return rainbow_color(argument)
    else:
        return False

arg = input("Enter a single letter for a rainbow color or a number for age calculation: ")
print(rainbow_or_age(arg))
```
